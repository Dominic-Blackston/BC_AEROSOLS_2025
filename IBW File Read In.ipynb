```python
import os
from igor2.binarywave import load
import numpy as np
import pandas as pd

# Set the path to the top-level folder
root_dir = "c:/Users/dblac/OneDrive/Desktop/OSU/BC_Research_Project/Data/CALNEX 2010/"

data_dict = {}

# Recursively walk through all subdirectories
for dirpath, _, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith(".ibw"):
            variable_name = os.path.splitext(filename)[0]
            file_path = os.path.join(dirpath, filename)

            try:
                with open(file_path, "rb") as f:
                    ibw_data = load(f)
                wave_data = ibw_data['wave']['wData']

                # Flatten if 1D array
                if wave_data.ndim == 1:
                    wave_data = wave_data.flatten()

                data_dict[variable_name] = wave_data

            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# Ensure AOCTimewave_all exists
if "AOCTimewave_all" not in data_dict:
    raise KeyError("AOCTimewave_all not found in the data. It is required to align time series.")

# Create time-indexed DataFrame
time_array = data_dict.pop("AOCTimewave_all")
df = pd.DataFrame(index=time_array)

# Add other variables
for key, values in data_dict.items():
    if len(values) == len(time_array):
        df[key] = values
    else:
        print(f"Skipping {key}: length mismatch with AOCTimewave_all")

# Display final DataFrame
print(df)
filename = 'CALNEX 2010'

df.to_csv('C:/Users/dblac/OneDrive/Desktop/OSU/BC_Research_Project/Data/CALNEX 2010/' + filename + '.csv')
```
