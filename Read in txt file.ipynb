```python
# Read 
import pandas as pd

file_path = "C:/Users/dblac/OneDrive/Desktop/OSU/BC_Research_Project/Data/TCAP/UHSAS/aaf.uhsasa010s.g1.tcap.20130207a.145705.a2.txt"


# Step 1: Open the file and read all lines
with open(file_path, 'r') as f:
    lines = f.readlines()

# Step 2: Find the line where actual data table starts (column headers)
for i, line in enumerate(lines):
    if line.startswith("Start_Time_(UTC)"):
        data_start_idx = i
        break

# Step 3: Read from that point onward
from io import StringIO
data_str = ''.join(lines[data_start_idx:])

# Step 4: Load into pandas
df = pd.read_csv(StringIO(data_str), parse_dates=['Start_Time_(UTC)'])

# Optional: Set datetime as index
df.set_index('Start_Time_(UTC)', inplace=True)

# âœ… Your data is now in a DataFrame
print(df.head())
```
